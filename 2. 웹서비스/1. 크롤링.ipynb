# **1. 크롤링**
크롤링(Crawling)은 웹사이트에서 데이터를 자동으로 수집하는 과정으로, 웹 크롤러라는 프로그램이 특정 웹 페이지를 방문해 그 안의 HTML 코드를 읽고 필요한 정보를 추출합니다. 이 과정은 일반적으로 웹 페이지를 탐색하면서 링크를 따라가며 반복되며, 주로 검색 엔진이 정보를 수집하거나 데이터 분석 목적으로 사용됩니다. 예를 들어, 뉴스 사이트의 제목과 본문을 수집해 정보를 분석하는 데 활용할 수 있습니다. 그러나 웹사이트의 규칙(robots.txt 파일)이나 법적 규제에 따라 크롤링이 제한될 수 있으므로 이를 준수하여야 합니다.

### 스크레이핑(Scraping): 크롤링 + 데이터를 추출하고 가공하는 행위

# **2. Basic English Speaking에서 75 Topics 크롤링**
* [Basic English Speaking](https://basicenglishspeaking.com/daily-english-conversation-topics/)

```python
import requests #소스코드를 받아오는 모듈
from bs4 import BeautifulSoup #requests를 통해 가져온 데이터를 분해해서 필요한 것만 꺼내는 것
```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
request = requests.get(site) #url 데이터를 이용하는 get방식
print(request) #200:정상적인 접속, 4XX: 페이지 없음
```

```python
print(request.text)
```

```python
soup = BeautifulSoup(request.text)
```

```python
div1 = soup.find('div', {'class': 'thrv-columns'})
print(div1)
```

```python
links = div1.find_all('a')
print(links)
```

```python
for link in links:
    print(link.text)
```

```python
subject = []
for link in links:
    subject.append(link.text)
print(subject)
print(len(subject))
```

```python
print(f'총 {len(subject)}개의 주제를 찾았습니다.')

for i in range(len(subject)):
    print(f'{i+1}. {subject[i]}')
```

# **3. 다음 뉴스기사 제목 크롤링**

```python
# https://v.daum.net/v/20241118174943073

# https://v.daum.net/v/20241118174013775

# https://v.daum.net/v/20241118142847486

def daum_news_title(news_id):
    url = f'https://v.daum.net/v/{news_id}'
    request = requests.get(url)
    # print(request)
    soup = BeautifulSoup(request.text)
    title = soup.find('h3', {'class':'tit_view'})
    if title:
        return title.text.strip()
    return '제목없음'
```

```python
daum_news_title(20241118174943073)
```

```python
daum_news_title(20241118174013775)
```

```python
daum_news_title(20241118142847486)
```

```python
daum_news_title(20241118193945321)
```

# **4. 벅스뮤직 차트 1~100위까지 크롤링**
* [벅스뮤직 차트](https://music.bugs.co.kr/chart)

```python
request = requests.get('https://music.bugs.co.kr/chart')
print(request)
```

```python
soup = BeautifulSoup(request.text)
```

```python
titles = soup.find_all('p', {'class':'title'})
artists = soup.find_all('p', {'class':'artist'})
```

```python
for i, (t, a) in enumerate(zip(titles, artists), 1):
    title = t.text.strip()
    artist = a.text.strip().split('\n')[0]
    print(f'{i}. {title} - {artist}')
```

# **5. 멜론차트 1~100위까지 크롤링**
* [멜론차트](https://www.melon.com/chart/index.htm)

```python
request = requests.get('https://www.melon.com/chart/index.htm')
print(request)
```

```python
# Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
request = requests.get('https://www.melon.com/chart/index.htm', headers=header)
print(request)
```

```python
soup = BeautifulSoup(request.text)
```

```python
titles = soup.find_all('div', {'class':'ellipsis rank01'})
artists = soup.find_all('div', {'class':'ellipsis rank02'})

print(titles)
print(artists)
```

```python
for i, (t, a) in enumerate(zip(titles, artists), 1):
    title = t.text.strip()
    artist = a.text.strip()
    print(f'{i}. {title} - {artist}')
```

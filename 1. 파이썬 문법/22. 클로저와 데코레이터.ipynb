# **1. 클로저**
클로저(Closure)는 내부 함수가 자신을 감싸고 있는 외부 함수의 변수에 접근할 수 있는 특성을 가지며, 외부 함수가 호출된 후에도 그 변수의 상태를 기억하고 사용할 수 있는 기능으로, 주로 데이터 은닉과 상태 유지에 활용됩니다.

```python
def mul2(n):
    return n*2
```

```python
print(mul2(10))
print(mul2(5))
```

```python
def mul5(n):
    return n*5
```

```python
print(mul5(5))
print(mul5(10))
```

```python
class Mul:
    def __init__(self, m):
        self.m = m

    def mul(self, n):
        return self.m * n
```

```python
mul2 = Mul(2)
print(mul2.mul(10))
print(mul2.mul(5))
```

```python
class Mul:
    def __init__(self, m):
        self.m = m

    def __call__(self, n):
        print('call 호출')
        return self.m*n
```

```python
mul2 = Mul(2)
print(mul2(10))
```

```python
mul5 = Mul(5)
print(mul5(10))
```

```python
# 클로저 사용하기
def mul(m):
    def wrapper(n):
        return m*n
    return wrapper
```

```python
mul2 = mul(2)
print(mul2(10))
```

```python
mul5 = mul(5)
print(mul5(10))
```

# **2. 데코레이터**
데코레이터(Decorator)는  기존 함수를 수정하거나 확장하는 함수로, 다른 함수를 인자로 받아 새로운 함수를 반환하여 원래 함수의 기능을 변경하거나 추가하는 데 사용되며, 주로 코드의 재사용성과 가독성을 높이는 데 활용됩니다.

```python
import time
```

```python
def func1(a, b):
    start = time.time()
    print('함수가 시작되었습니다')
    result = a + b
    end = time.time()
    print(f'함수 수행시간: {end - start}')
    return result
```

```python
print(func1(10, 3))
```

```python
def func2(a, b):
    start = time.time()
    print('함수가 시작되었습니다')
    result = a * b
    end = time.time()
    print(f'함수 수행시간: {end - start}')
    return result
```

```python
print(func2(10, 3))
```

```python
def func1(a, b):
    result = a + b
    return result

def func2(a, b):
    result = a * b
    return result
```

```python
def elapsed(func):
    def wrapper(a, b):
        start = time.time()
        print('함수가 시작되었습니다')
        result = func(a, b)
        end = time.time()
        print(f'함수 수행시간: {end - start}')
        return result
    return wrapper
```

```python
deco1 = elapsed(func1)
result = deco1(10, 3)
print(result)
```

```python
deco2 = elapsed(func2)
result = deco2(10, 3)
print(result)
```

```python
@elapsed
def func1(a, b):
    result = a + b
    return result

print(func1(10, 3))
```

```python
@elapsed
def func2(a, b):
    result = a * b
    return result

print(func2(10, 3))
```

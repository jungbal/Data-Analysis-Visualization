# **1. 사용자 정의 함수**
사용자 정의 함수란 사용자가 특정 작업을 수행하기 위해 직접 작성한 함수를 의미합니다. 파이썬에는 많은 내장 함수들이 있지만, 때로는 우리의 요구사항에 맞게 동작하는 함수를 직접 만들어야 할 때가 있습니다. 이때 사용자 정의 함수를 작성하게 됩니다.

```
def 함수명(매개변수1, 매개변수2, ...):
    # 함수 내용
    return 결과값
```

# **3. 매개변수가 있고 반환 값이 없는 함수**

````python
def func2(num):
  print(f'입력받은 숫자: {num}')
````

````python
func2(10)
func2(4)
````

````python
def func3(start, end):
  sum = 0
  for i in range(start, end + 1):
    sum += i
  print(f'{start}부터 {end}까지의 합: {sum}')
````

````python
func3(1, 10)
````

# **4. 반환 값이 있는 함수**

````python
def func4():
  print('안녕하세요 파이썬')

func4() # 함수실행
temp = func4() # 함수 실행 후 return 값을 temp에 저장(return 값이 없어서 none이 저장)
print(temp)

temp = func4 # 함수의 메모리 주소를 temp에 저장
print(temp)
temp()
````

````python
def func5():
  return '🎁'

temp = func5()
print(temp)
print(f'temp에 저장된 값: {temp}')
````

````python
def func6(num1, num2):
  sum = num1 + num2
  return sum

print(func6(10, 5))
````

````python
temp = func6(1, 2)
print(temp)
````

# **5. None의 특징**
None은 파이썬에서 특별한 값으로, 아무런 값이 없음을 표현하는 데 사용됩니다. 다른 언어에서의 null 또는 nil과 유사한 개념입니다. None은 파이썬의 내장 상수이며, 그 자체로 데이터 타입이 NoneType입니다. 모든 None은 동일하므로, 두 개의 None 값을 비교할 때 항상 True를 반환합니다.

````python
variable = None
````

````python
def func():
  pass
````

````python
def func(x):
  if x > 0:
    return x
````

````python
print(func(1))
print(func(-1))
````

````python
def func(message=None):
    if message is None:
        print("Hello!")
    else:
        print(message)
````

````python
func('Hi')
func()
````

````python
li1 = [1, 2, 3, 4, 5]
li2 = [1, 2, 3, 4, 5]

# 객체 비교
result1 = li1 is li2
# 값 비교
result2 = li1 == li2

print(result1)
print(result2)
````

````python
# 값의 존재 여부 확인
def get_data_from_database():
    pass

data = get_data_from_database()
print(data)
if data is None:
    print('데이터를 수신하지 못함!')
else:
    print('데이터를 수신받음!')
````

# **7. 가변 매개변수**
함수를 호출할 때 *를 사용하면 시퀀스(리스트, 튜플 등)의 요소를 개별적인 위치 인자로 풀어서 전달할 수 있습니다.

````python
def func7(*args):
    return args
````

````python
print(func7())
print(func7(10))
print(func7(10, 30, 50))
````

````python
def func8(a, b, c):
  return a + b + c
````

````python
numbers = [1, 2, 3]
func8(*numbers)
````

# **8. 키워드 매개변수**
키워드 매개변수는 일반적으로 기본값이 설정된 매개변수와 함께 사용됩니다. 함수의 매개변수에 기본값을 설정하면, 함수를 호출할 때 해당 매개변수를 생략할 수 있습니다.

````python
def func9(id, name, age):
    print(f'아이디: {id}')
    print(f'이름: {name}')
    print(f'나이: {age}')
````

````python
func9('apple', '김사과', 20)
func9(age=30, id='orange', name='오렌지')
````

````python
# 매개변수명과 딕셔너리의 키가 같아야함
# 딕셔너리의 키는 반드시 문자열 형태
dic1 = {'age':25, 'id':'banana', 'name':'반하나'}
func9(**dic1)
````

````python
func9(*dic1)
````

# **여러개의 반환 값**

````python
def func10(num1=0, num2=0):
    return num1+num2, num1-num2, num1*num2, num1/num2
````

````python
result = func10(10, 3)
print(result)
````

````python
result1, result2, result3, result4 = func10(10, 3)
print(result1)
print(result2)
print(result3)
print(result4)
````

````python
_, _, result3, _ = func10(10, 3)
print(f'곱셈:{result3}')
````

````python
li1 = [10, 20, 30, 40, 50]

for _, v in enumerate(li1):
    print(f'값: {v}')
````

# **1. 튜플**
여러 가지 데이터 타입의 값을 저장할 수 있는, 변경 불가능(immutable)한 순서가 있는 컬렉션입니다.

### 1. 생성

튜플은 괄호 ()를 사용하여 생성하고, 항목들은 쉼표 ,로 구분됩니다.

```python
tu1 = ()
print(tu1)
print(type(tu1))
```

```python
tu1 = (1,) # 요소가 1개일 경우 끝에 콤마를 꼭! 붙여야 함
print(tu1)
print(type(tu1))
```

```python
tu2 = (1, 3, 5, 7)
print(tu2)
print(type(tu2))
```

```python
tu3 = tuple([1, 3, 5, 7])
print(tu3)
print(type(tu3))

li3 = list(tu3)
print(tu3)
print(li3)
print(type(li3))

tu4 = 1, 2, 3, 4
print(tu4)
print(type(tu4))
```

```python
# 패킹과 언패킹
num1 = 10
num2 = 20
num3 = 30
tu4 = num1, num2, num3
print(tu4)
print(type(tu4))

result1, result2, result3 = tu4
print(result1)
print(result2)
print(result3)
print(type(result1))
print(type(result2))
print(type(result3))
```

```python
tu5 = ('apple', 'banana', ('🍎', '🍌'))
print(tu5)
print(type(tu5))
print(tu5[0])
print(type(tu5[0]))
print(tu5[2][0])
print(type(tu5[2][0]))
```

```python
tu5 = ('apple', 'banana', ('🍎', '🍌'))
print(tu5)
print(type(tu5))
tu5 = 'orange'
print(tu5)
print(type(tu5))
```

```python
tu5 = ('apple', 'banana', ['🍎', '🍌'])
print(tu5)
print(type(tu5))
print(tu5[2])
print(type(tu5[2]))
tu5[2][0] = '🍓'
print(tu5)
# tu5[2] = '🍉' #TypeError: 'tuple' object does not support item assignment
```

```python
# 튜플의 슬라이싱은 결과도 튜플로 반환
tu6 = (1, 2, 'apple', 'banana')
print(tu6[0])
print(tu6[-1])
print(tu6[1:])
print(tu6[1:3])
```

```python
tu7 = (10, 20, 30)
tu8 = (40, 50, 60)

print(tu7 + tu8)
print(tu7 * 3)

tu7 = tu7 + (40, 50, 60)
print(tu7)
```

```python
print(2 in [2, 4, 5, 6, 7, 8])
print(2 in (2, 4, 5, 6, 7, 8))
print(3 in (2, 4, 5, 6, 7, 8))
```

```python
tu9 = (10, 20, 30, 40, 100, 300, 200, 90)
# tu10.sort() -> 튜플은 sort 메서드가 존재하지 않음
result = sorted(tu9, reverse = True) # 내림차순하여 리스트로 반환
print(result)
print(type(result))
print(tu9) # 변경되지 않음
result = tuple(result)
print(result)
```

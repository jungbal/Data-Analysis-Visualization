# **5. 컬렉션 타입**
여러 개의 데이터 항목을 하나의 단위로 관리할 수 있게 해주는 데이터 구조를 의미합니다. 이것은 여러 개의 데이터를 하나의 변수에 저장하고, 관리할 수 있게 해줍니다. 파이썬에서는 리스트, 튜플, 세트, 딕셔너리 등이 기본적인 컬렉션 타입에 속합니다.

# **2. 리스트**
여러 값들을 하나의 변수에 저장하고 관리할 수 있게 해주는 순차적인 자료구조입니다.

```python
list1 = [1, 3, 5, 7, 9]
print(list1)
print(type(list1))

list2 = list([1, 3, 5, 7, 9])
print(list2)
print(type(list2))

list3 = ['오벅벅', '진미미', '조대감', '천주달']
print(list3)

list4 = ['오벅벅', 1, 3.14, True]
print(list4)
print(type(list4))
print(list4[1])
print(type(list4[1]))
print(list4[2])
print(list4[3])
# print(list4[4]) #IndexError: list index out of range -> 존재하지 않는 인덱스 번호 호출
```

```python
list1 = [1, 3, 5, 7, 9]
print(list1)
print(list1[0])
print(list1[-1])
print(list1[0] + list1 [-1])

list2 = [1, 2, '파이썬', ['오벅벅', '진미미']]
print(list2)
print(type(list2))
print(list2[1])
print(type(list2[1]))
print(list2[3])
print(type(list2[3]))
print(type(list2[3][0]))
```

```python
list1 = [10, 20, 30, 40, 50]
print(list1)
print(list1[0:3])

list2 = list1
print(list2)

list2[0] = 100
print(list2)
print(list1)

list3 = [1, 2, 3, ['오벅벅', '진미미', '조대감', '천주달', ['🐱', '🦁', '🐯', '🐶']]]
print(list3)
print(list3[2:3])
print(list3[3][:2])
print(list3[3][2])
print(list3[-1][-1][-2])
```

```python
list4 = [10, 20, 30, 40, 50]
# 슬라이싱을 이용하여 요소를 추가한 경우 리스트에 데이터만 포함
list4[1:2] = ['😁','😂','😎','😍']
print(list4)

list4 = [10, 20, 30, 40, 50]
# 인덱싱을 이용하여 요소를 추가한 경우 리스트 안에 리스트를 만들고 포
list4[1] = ['😁','😂','😎','😍']
print(list4)

list4 = [10, 20, 30, 40, 50]
print(list4[1:3])
list4[1:3] = [] # 빈 리스트를 슬라이싱을 통해 저장하면 해당 요소가 삭제됨
print(list4)

list4 = [10, 20, 30, 40, 50]
print(list4[1:2])
print(list4[1])
list4[1] = []
print(list4)

list4 = [10, 20, 30, 40, 50]
list4[1:2] = []
print(list4)

list4 = [10, 20, 30, 40, 50]
print(list4[1])
del list4[1]
print(list4)
```

```python
list5 = [10, 20, 30]
list6 = [40, 50, 60]
print(list5 + list6)
print(list6 + list5)

list5 = [10, 20, 30]
list6 = list5 + [40, 50, 60]
print(list5)
print(list6)

list6 = [40, 50, 60]
list6 += [40, 50, 60]
print(list6)

list6 = [40, 50, 60]
print(list6 * 3)
```

# **3. 함수와 메서드**

```python
# len(): 객체의 길이를 반환하는 파이썬의 기본 내장 함수
list6 = [10, 20, 30]
print(len(list6))
```

```python
# append(): 리스트 요소의 끝에 새로운 요소를 추가
list6 = [10, 20, 30]
print(list6)

list6.append(40)
print(list6)

# list6.append(50, 60) -> TypeError
list6.append([50, 60])
print(list6)
```

```python
# extend(): 리스트 요소의 끝에 새로운 여러 요소를 추가
li6 = [10, 20, 30]
print(li6)

# li6.extend(100) #TypeError -> list 형태로만 가능
li6.extend([100])
print(li6)
li6.extend([200, 300, 400])
print(li6)
```

```python
# pop(): 리스트 마지막 요소를 삭제하고 삭제된 요소를 반환
li6 = [10, 20, 30]
print(li6)

# print(li6.pop()) # -> 주석문을 실행하는 순간 30 삭제 10, 20이 남고 아래 문장에서 변수에 저장할 때는 맨 끝 요소가 20이기 때문에 20이 temp에 저장되고 li6에는 10이 남고 li6.pop()을 출력하면 10이 출력되고 li6을 출력하면 공백이 출력
temp = li6.pop()
print(li6)
print(temp)
print(li6.pop())
print(li6)
```

```python
# index(): 리스트에서 특정 요소의 값 인덱스를 반환
li6 = [10, 20, 30]
print(li6)

print(li6.index(30))
# print(li6.index(100)) -> ValueError: 100 is not in list 반환할 값이 없음
```

```python
# reverse(): 리스트의 요소들의 순서를 반대로 설정
li7 = [100, 50, 70, 60, 20]
print(li7)

li7.reverse()
print(li7)
```

```python
# 슬라이싱을 사용하여 리스트의 요소들의 순서를 반대로 설정하는 방법
li7 = [100, 50, 70, 60, 20]
print(li7[:]) # [0:5]
print(li7[0:5])
print(li7[::])    # step이 생략
print(li7[0:5:1]) # step의 기본값은 1
print(li7[0:5:2]) # step을 2로 설정

li7 = [100, 50, 70, 60, 20]
print(li7[::-1]) # step을 -1로 설정 -> start 값의 기본값이 -1, stop의 기본 값이 0으로 설정 -> reverse와 같은 값으로 출력

print(li7[1::-1])
print(li7[:0:-1])
```

```python
# sort(): 리스트의 요소를 오름차순으로 정렬
li7 = [100, 50, 70, 60, 20]
print(li7)

li7.sort()
print(li7)
# sort(reverse=True): 리스트의 요소를 내림차순으로 정렬
li7.sort(reverse=True)
print(li7)

li8 = ['Apple', 'apple', 'orange', 'banana', 'melon']
li8.sort()
print(li8)

li9 = ['김사과', '오렌지', '반하나', '이메론', '배애리']
li9.sort()
print(li9)
```

```python
# sorted(): 모든 요소를 정렬한 후 반환해주는 함수
li7 = [100, 50, 70, 60, 20]
print(li7)

result = sorted(li7)
print(li7)
print(result)
```

```python
# count(): 리스트에서 특정 요소의 갯수를 반환
li9 = [10, 20, 30, 50, 20, 40, 30, 20]
print(li9)

print(li9.count(20))
print(li9.count(100))
```
